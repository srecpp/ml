```python
import matplotlib.pyplot as plt
a = [5, 2, 6, 4, 3]
b = [4, 7, 0.5, 15, 10, 8, 16, 21]
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(a, marker='o', color='r')
axs[0, 0].set_title('Dataset A')
axs[0, 1].plot(b, marker='s', color='g')
axs[0, 1].set_title('Dataset B')
axs[1, 0].axis('off')
axs[1, 1].axis('off')
plt.tight_layout()
plt.show()
```
